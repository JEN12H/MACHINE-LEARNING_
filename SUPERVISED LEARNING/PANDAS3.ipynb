```python
import pandas as pd
india_weather = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[22,45,30],
    "humidity":[80,60,70]
})
india_weather
```

```python
us_weather = pd.DataFrame ({
    "city":["newyork","chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
 df= pd.concat([india_weather,us_weather])
df
```

```python
df= pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
df= pd.concat([india_weather,us_weather],keys =["INDIA","US"])
df
```

```python
df.loc["INDIA"]
```

```python
temperature_df =pd.DataFrame ({
    "city":["newyork","chicago","orlando"],
    "temperature":[21,14,35]
},index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame ({
    "city":["chicago","newyork","orlando"],
    "humidity":[68,65,75]
},index=[1,0,2])
windspeed_df
```

```python
df= pd.concat([temperature_df,windspeed_df],keys =["INDIA","US"],axis =1)
df
```

```python
s = pd.Series(["Humid","Dry","Rain"],name="event")
s
```

```python
df = pd.concat([temperature_df,s],axis = 1)
df
```

```python
df1 =pd.DataFrame ({
    "city":["newyork","chicago","balitmore"],
    "temperature":[21,14,35],
})
df1
```

```python
df2 = pd.DataFrame ({
    "city":["chicago","newyork","san fransisco"],
    "humidity":[68,65,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city")
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how ="outer",indicator = True)
df3
```

```python
df3 = pd.merge(df1,df2,on="city",how ="inner")
df3
```
