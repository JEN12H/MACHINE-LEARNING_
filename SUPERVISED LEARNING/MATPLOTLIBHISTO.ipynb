```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGLE','AMAZON','MICROSOFT','FACEBOOK']
revenue = [90,136,89,37]
profit =[40,2,34,12]
```

```python
xpos =np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel("REVENUE(BN$)")
plt.title("DATA")
plt.bar(xpos,revenue,label="Revenue")
plt.bar(xpos,profit,label="Profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel("REVENUE(BN$)")
plt.title("DATA")
plt.barh(xpos,revenue,label="Revenue")
plt.barh(xpos,profit,label="Profit")
plt.legend()
```
