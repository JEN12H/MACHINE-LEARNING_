# LOGISTIC REGRESSION - CLASSIFICATION


```python
#USES SIGMOID FUNCTION : sigmoid = 1/ 1+e*(-mx+b) - IT PROVIDES VALUE BETWEEN 0 AND 1 
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv("insurance_data.csv")
df.head()
df.tail()
```

```python
plt.scatter(df.age,df.bought_insurance,color='blue')
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df[['age']],df.bought_insurance,test_size = 0.1,train_size = 0.9)
```

```python
X_test
```

```python
X_train
```

```python
from  sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test) # MEANS YOUNGER AGE IN MY MODEL(19,0,11) ARE NOT GOING TO BUY INSURANCE (0,0,0)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test)
```
